If the initial state of the system (time=0) is "bear," what will be the probability distribution at time=3?  What about time=infinity (the steady state probability distribution).

```python
import numpy as np
import pandas as pd
```

```python
market = [[.9,.075,.025],
[.15,.8,.05],
[.25,.25,.5]]
```

```python
def bull(probability):
    if probability < .025:
        return (3,"stagnant")
    elif probability < .01:
        return (2,"bear")
    elif probability < 1:
        return (1,"bull")
```

```python
def bear(probability):
    if probability < .15:
        return (1,"bull")
    elif probability < .2:
        return (3,"stagnant")
    elif probability < 1:
        return (2,"bear")
```

```python
def stagnant(probability):
    if probability < .25:
        return (1,"bull")
    elif probability < .5:
        return (3,"bear")
    elif probability < 1:
        return (2,"stagnant")
```

```python
runs = []
for x in range(0,100):
    state = 1
    run = [1]
    for i in range(0,3):
        probability = np.random.random(1)
        if(state == 1):
            state, stateName = bull(probability)
        elif(state == 2):
            state, stateName = bear(probability)
        elif(state == 3):
            state, stateName = stagnant(probability)
        run.append(state)
    runs.append(run)
```
