##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
formattedDF = pd.melt(df, id_vars=df.columns.tolist()[0:7], value_vars=df.columns.tolist()[7:81], var_name='Week', value_name='Rank')
formattedDF
formattedDF = formattedDF[formattedDF.Rank.notnull()]

import re
non_decimal = re.compile(r'[^\d]+')

formattedDF.Week = [int(non_decimal.sub('', x)) for x in formattedDF.Week]

import datetime

def formatDate(date,weeks):
    u = datetime.datetime.strptime(date,"%Y-%m-%d")
    d = datetime.timedelta(days=weeks*7)
    t = u + d
    return t.strftime("%Y-%m-%d")

def get_sec(s):
    l = s.split(':')
    return int(l[0]) * 60 + int(l[1])

def weeks_between(d1, d2):
    d1 = datetime.datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days/7)

formattedDF["WeekDate"] = [formatDate(x,int(y)) for x,y in zip(formattedDF["Debut Date"],formattedDF.Week)]

formattedDF["InvertedRank"] = [101 - x for x in formattedDF.Rank]

formattedDF["Track Length in Seconds"] = [ get_sec(x) for x in formattedDF.Length]

formattedDF["Time to Peak"] = [ weeks_between(x,y) for x,y in zip(formattedDF["Debut Date"],formattedDF["Peak Date"])]

formattedDF.to_csv('../assets/formattedBillboard.csv')
```
