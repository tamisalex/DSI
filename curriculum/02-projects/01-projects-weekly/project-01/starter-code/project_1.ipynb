## Step 4: Visualize the data

##### 14. Using MatPlotLib and PyPlot, plot the distribution of the Rate using histograms.

```python
import seaborn as sns

%matplotlib inline

def distribution_plotter(column, data):
    sns.set(rc={"figure.figsize": (10, 7)})
    sns.set_style("white")
    dist = sns.distplot(data, hist_kws={'alpha':0.2}, kde_kws={'linewidth':5})
    dist.set_title('Distribution of ' + column + '\n', fontsize=16)
    
distribution_plotter(data.columns.values[1],data.iloc[:,1:2])
```

##### 15. Plot the Math distribution

```python
distribution_plotter(data.columns.values[3],data.ix[1:,"Math"])
```

##### 16. Plot the Verbal distribution

```python
distribution_plotter(data.columns.values[2],data.ix[1:,"Verbal"])
```

##### 19. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
import matplotlib.pyplot as plt

rate = data.ix[1:,"Rate"].values
x = data.ix[1:,"Verbal"].values
y = data.ix[1:,"Math"].values

fig = plt.figure(figsize=(12,8))
plt.scatter(np.array(x), np.array(rate))
#plt.axis([450,650,450,650])

axes = fig.gca()
axes.set_title('Verbal vs Rate scores\n')
axes.set_xlabel('Math Scores')
axes.set_ylabel('Rate')
#axes.set_autoscale_on(False)

plt.show

fig2 = plt.figure(figsize=(12,8))
plt.scatter(np.array(y), np.array(rate))
#plt.axis([450,650,450,650])

axes2 = fig.gca()
axes2.set_title('Math vs Rate scores\n')
axes2.set_xlabel('Verbal Scores')
axes2.set_ylabel('Rate')
#axes.set_autoscale_on(False)

plt.show()
```
